```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(1, 20)
```

```python
#Create figure and axes
fig, ax = plt.subplots(figsize=(18.5, 10.5))
ax.plot(n ,np.sqrt(n), label='Sqrt', marker='x', color='r') # plot some data on axes
ax.plot(n, n, label='linear', marker='x', color='g')
ax.plot(n, n + np.sqrt(n), label='n plus sqrt',marker='o', color='b')
ax.set_xlabel('x label') # Add x label to axes
ax.set_ylabel('y label') # Add y label to axes
ax.set_title('Function comparision')
ax.legend() # Add a legend
```

```python
#Create figure and axes
fig, ax = plt.subplots(figsize=(18.5, 10.5))
ax.plot(n, np.log(n), label='log',marker='x', color='r')
ax.plot(n, n*np.log(n), label='nlogn', marker='x', color='g')
ax.set_xlabel('x label') # Add x label to axes
ax.set_ylabel('y label') # Add y label to axes
ax.set_title('Function comparision')
ax.legend() # Add a legend
```

```python
fig, ax = plt.subplots(figsize=(18.5, 10.5))
ax.plot(n, 2**n, label='2**n',marker='x', color='r')
ax.plot(n, n**2, label='n**2', marker='o', color='g')
#ax.plot(n, n*np.log(n), label='nlogn', marker='x', color='g')
ax.set_xlabel('x label') # Add x label to axes
ax.set_ylabel('y label') # Add y label to axes
ax.set_title('Function comparision')
ax.legend() # Add a legend
```

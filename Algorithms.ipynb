### Characterstics of algorithm
- Input -- 0 or more input
- Output -- generate some output atleast 1 output
- Definiteness -- Every statement should be unambiguous
- Finiteness -- Must terminate at some point
- Effectiveness -- Every statement should serve purpose


### How to write an algorithm
Algorithm swap(a , b)
```python
def swap(a, b):
    temp = a
    a = b
    b = temp
```

### How to analyse algorithm
* Time -- How much time it takes to execute algorithm. Time Function
* Space -- How much memory space it is going to consume. Space complexity
* Network consumption -- How much data transfer is done.
* Power consumption -- How much power it is consuming
* CPU registers consumption


**Every statement one unit of time** Like swap function defined time function is $f(n) = 3$. It is constant so $O(n) = O(1)$ To analyse in more depeth we need to consider how many operations are done in each statement.

**Every variable defined is one unit of space** For example swap function defined above space complexity is $S(n) = 3$, there are three variables used $a, b, temp$. Space complexity is constant which indicates $O(n) = O(1)$.

**Algorithm to get sum of all elements in an array.
sum(A, n)**
```c
{
int s = 0; /* -- 1 operation */
for (i = 0; i <n; i++) /* -- n + 1 operation */
{
    s = s + A[i]; /* --n operation */
 }
return s; /* -- 1 operation */
}
```
Time function is $f(n) = 2n + 3$  Order of $O(n)$

Space complexity  Varibales defined are $A, n, s, i$ size of $A$ is $n$ and size of $n , s, i$ is $1$ which gives $S(n) = n + 3$
